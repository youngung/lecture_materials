Newton Raphson

```python
def f(x):
    return exp(x)+2*cos(x)-3
def fp(x):
    return exp(x)-2*sin(x)
def gen_line(x):
    def l(xr):        
        return f(x)+fp(x)*(xr-x)
    return l
```

```python
x=np.linspace(-1,1)
plot(x,f(x),label='func')


x0=0.5
plot(x0,f(x0),'s',label='Init')
fl=gen_line(x0)
plot(x,fl(x),'--',label='1st line')


x1=x0-f(x0)/fp(x0)
plot(x1,f(x1),'x',label='2nd')
fl=gen_line(x1)
plot(x,fl(x),'--',label='2nd line')


x2=x1-f(x1)/fp(x1)
plot(x2,f(x2),'s',label='3rd')
fl=gen_line(x2)
plot(x,fl(x),'--',label='3rd line')


grid()
legend()
```
